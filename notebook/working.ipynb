```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings_with_books = ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.head()
```

```python
number_ratings = ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_ratings.head()
```

```python
number_ratings.rename(columns={'rating':'num_of_ratings'}, inplace=True)
```

```python
final_rating = ratings_with_books.merge(number_ratings, on='title')
```

```python
final_rating.head()
```

```python
final_rating = final_rating[final_rating['num_of_ratings'] >= 50]
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['title','user_id'], inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot.head()
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[237,:]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1), n_neighbors=10 )
```

```python
# Model testing
def recommend_books(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=11)

    for i in range(len(suggestion)):
        books = book_pivot.index[suggestion[i]]
        for j in books:
            if j == book_name:
                print(f"You searched '{book_name}'\n")
                print("The suggestion books are: \n")
            else:
                print(j)
```

```python
book_name = "A Bend in the Road"
recommend_books(book_name)
```
